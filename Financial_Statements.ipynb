# **Financial Analysis of Selected Public Companies**
In this project, financial statement data from various publicly traded companies was analyzed to derive meaningful insights into their financial performance. The analysis focuses on key metrics such as revenue, net income, market capitalization, and profitability ratios. By comparing companies from different sectors (e.g., fintech, technology, and e-commerce), the study highlights trends, strengths, and potential risks through visualizations and metric-based evaluations. The goal is to demonstrate analytical thinking, data visualization skills, and business understanding in the context of financial data.

In this project, the "Financial Statements of Major Companies (2009–2023)" dataset from Kaggle was used.

### **Column Descriptions:**

* **Year**: The year in which the financial data was recorded.
* **Company**: The name of the company.
* **Category**: The industry or sector to which the company belongs.
* **Market Cap (in B USD)**: Total market value of the company’s outstanding shares, in billion USD.
* **Revenue**: Total income generated by the company from its operations.
* **Gross Profit**: Revenue minus the cost of goods sold (COGS).
* **Net Income**: The company’s total profit after all expenses, taxes, and costs.
* **Earning Per Share (EPS)**: Portion of a company’s profit allocated to each outstanding share.
* **EBITDA**: Earnings before interest, taxes, depreciation, and amortization.
* **Share Holder Equity**: The net value of a company owned by its shareholders.
* **Cash Flow from Operating**: Cash generated from core business activities.
* **Cash Flow from Investing**: Cash used in or generated from investment activities like asset purchases.
* **Cash Flow from Financial Activities**: Cash from financing activities such as debt or equity issuance.
* **Current Ratio**: A liquidity ratio measuring the ability to pay short-term obligations.
* **Debt/Equity Ratio**: A measure of a company’s financial leverage, calculated by dividing total liabilities by shareholder equity.
* **ROE** *(Return on Equity)*: Net income as a percentage of shareholder equity, indicating profitability.
* **ROA** *(Return on Assets)*: Net income relative to total assets, measuring how effectively assets generate profit.
* **ROI** *(Return on Investment)*: Measures the gain or loss generated on an investment relative to its cost.
* **Net Profit Margin**: Percentage of revenue remaining after all expenses have been deducted.
* **Free Cash Flow per Share**: Cash remaining after capital expenditures, divided by the number of shares.
* **Return on Tangible Equity**: Profitability ratio measuring returns on physical (non-intangible) equity.
* **Number of Employees**: Total number of employees working at the company.
* **Inflation Rate (in US)**: The annual inflation rate in the United States for the given year.

## Market Cap vs EPS Comparison
**Why Compare Market Capitalization and Earnings Per Share (EPS)?**

Market Capitalization represents the total market value of a company's outstanding shares and reflects investor perception of the company's worth. Earnings Per Share (EPS), on the other hand, indicates how much profit is allocated to each outstanding share of common stock.

By comparing Market Cap and EPS over time:


*  We can observe whether a company’s valuation growth is supported by its actual profitability.
*  It helps identify discrepancies where market value may rise despite stagnant or declining earnings.
*   This comparison also highlights investor confidence relative to a company's financial performance.

## Gross Profit Margin vs Net Profit Margin
Comparing Gross Profit Margin (GPM) and Net Profit Margin (NPM) helps to understand how operating expenses and other costs affect a company's overall profitability. While a high GPM indicates strong core profitability, a significantly lower NPM may reveal high operating or financial costs that are reducing net income. This analysis sheds light on the effectiveness of cost management beyond direct production costs.

---



First of all, since there is no “gross_profit_margin” column in our data, we can calculate the gross profit margin as follows.

General formula:
$$
\text{Gross Profit Margin} = \frac{\text{Gross Profit}}{\text{Revenue}} \times 100
$$


```python
df['gross_profit_margin'] = (df['gross_profit'] / df['revenue']) * 100
```

To see the impact of the company's operating expenses on margins, we can subtract the net profit margin from the gross profit margin. If this difference is high, it indicates that the company's operating expenses (tax, interest, depreciation, etc.) are significantly eating into gross profit.

```python
df['margin_difference'] = df['gross_profit_margin'] - df['net_profit_margin']
```

```python
import plotly.express as px

fig = px.histogram(df, x='company',  y='margin_difference', color='year', barmode='group',
             title='Difference Between Gross Profit Margin and Net Profit Margin')
fig.update_layout(
    xaxis_title='Company',
    yaxis_title='Margin Difference (%)',
)
fig.show()
```

## Impact of inflation rate on Financial Performance
This analysis examines how the rate of inflation affects key financial metrics such as net income, EBITDA, free cash flow per share, and profit margins. Inflation can erode purchasing power, increase operating costs, and affect both revenues and profitability. Understanding this relationship helps to assess whether companies are resilient in an inflationary environment or whether their performance will suffer.

```python
df[['inflation_rate', 'net_income', 'ebitda', 'free_cash_flow_per_share', 'net_profit_margin']].corr()['inflation_rate']
```

```python
corr_cols = ['inflation_rate', 'net_income', 'ebitda',
             'free_cash_flow_per_share', 'net_profit_margin']


corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Between Inflation and Financial Metrics')
plt.show()
```

The relationship between **net income** (net profit) and **the inflation rate** helps us understand how the company's profitability is affected by inflation.

**Positive Correlation:** If net profit increases as inflation increases, the company may be able to pass on price increases to customers (e.g. firms with strong pricing power).

**Negative Correlation:** If net profit decreases when inflation increases, this may indicate that the company is having difficulty passing on increased costs to customers or that its operational costs have increased significantly with inflation.

**Weak or close to zero correlation:** Inflation may not have a significant impact on the company's net profit, which could be a sign of operational stability.

This analysis provides investors with an opportunity to assess the company's resilience to economic fluctuations.

```python
companies = df['company'].unique()
years = df['year'].unique()

# Only 1 trace per company
traces = []
for company in companies:
    company_df = df[df['company'] == company]
    traces.append(go.Scatter(
        x=company_df['inflation_rate'],
        y=company_df['net_income'],
        mode='markers',
        marker=dict(
            size=12,
            color=company_df['year'],  # color by year
            colorscale='Viridis',
            showscale=True,
            colorbar=dict(title='Year')
        ),
        name=company,
        hovertemplate='Year: %{marker.color}<br>Inflation: %{x}<br>Net Income: %{y}',
        visible=(company == companies[0])  # Only the first company is visible at first
    ))

# Dropdown buttons
dropdown_buttons = [
    {
        'label': company,
        'method': 'update',
        'args': [
            {'visible': [c == company for c in companies]},
            {'title': f'Inflation Rate vs Net Income for {company}'}
        ]
    } for company in companies
]

# Layout
fig = go.Figure(data=traces)
fig.update_layout(
    title=f'Inflation Rate vs Net Income for {companies[0]}',
    xaxis_title='Inflation Rate',
    yaxis_title='Net Income',
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction='down',
            showactive=True,
            x=0.5,
            xanchor='center',
            y=1.15,
            yanchor='top'
        )
    ]
)

fig.show()
```
